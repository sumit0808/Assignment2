```python
#Write a python program to scrap book name, author name, genre and book review of any 5 books from
#‘www.bookpage.com’
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get("https://bookpage.com/reviews")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
book_name=soup.find_all('h4',class_="italic")
book_name
```

```python
books=[]
for index, name in enumerate(book_name):
    if index <5:
        books.append(name.text.replace("★","").strip())
books
```

```python
author_name=soup.find_all('p',class_="sans bold")
author_name
```

```python
authors = []
for author in author_name:
    authors.append(author.text.strip())

authors[:5]
```

```python
genre=soup.find_all('p',class_="genre-links hidden-phone")
# genre
```

```python
genres = []
for gen in genre:
    genres.append(gen.text.strip().split()[0])
    
genres
```

```python
book_review=soup.find_all('div',class_="read-full")
book_review

reviews = []
for review in book_review:
    href = review.find('a', href=True)['href']
    url = "https://bookpage.com"+href
    page = requests.get(url)
    
    soup1 = BeautifulSoup(page.content)
    data = soup1.find_all('div',class_="article-body")
    
    review_article = ""
    
    for article in data:
        for row in article.find_all('p'):
            review_article+=row.text
    reviews.append(review_article.strip())
    
reviews[:5]
```

```python
import pandas as pd

df=pd.DataFrame()
df['Book Name']=books[:5]
df['Auther name']=authors[:5]
df['genre'] = genres[:5]
df['review']=reviews[:5]

df
```

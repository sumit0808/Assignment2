```python
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
page.text
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
li=["h1","h2","h3","h4","h5","h6"]
header=soup.find_all(li)
header
```

```python
for i in header:
    print(i.text)
    print()
```
